In this excercise, we will implement the pocket algorithm for the binary classification.

```python
import numpy as np

%matplotlib inline
import sklearn.datasets
import matplotlib.pyplot as plt
import matplotlib
```

```python
# 1. 데이터 생성
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_moons.html

np.random.seed(0)

X, y = sklearn.datasets.make_moons(200, noise=0.1) # noise를 줄이면 선처럼 분포가 모이게 바뀜 1로 주면 너무 퍼져서 달모양 아닌 것처럼

# 데이터 크기 및 샘플 확인
print(X[:5])
print(y[:5])
print(X.shape, y.shape)
```

```python
# 2. 데이터 시각화
plt.scatter(X[:,0], X[:,1],
           s=40,
           c=y,
           cmap=plt.cm.Spectral)
```

```python
# 3. bias 설정

# X 의 모든 데이터에 bias 로 활용될, x_0 = 1 을 추가
oneVector = np.ones((X.shape[0],1)) # 200x1짜리 1로 만들어진 벡터 만든다
print(oneVector.shape)

# 기존에 x_1, x_2만 존재하는 X 에 x_0 을 붙임
X_train = np.concatenate((oneVector, X), axis=1) # 각 데이터에 X0를 추가해줌 200x2를 200x3으로 바꾼 것
print(X_train.shape)
print(X_train[:5]) 
```

```python
# 4. 파라미터 설정

# 학습률 (learning rate) 설정
# -> 각 iteration 에서 가중치의 갱신 정도를 결정
learningRate = 0.01

# 파라미터 (parameter) 또는 가중치 (weights) 설정
# -> [w_0, w_1, w_2] 을 의미
weights = np.random.rand(3,1)

# 최대 오분류 샘플 수 설정
# -> pocket algorithm 에서는 모든 데이터에 대해 평가 (evaluation) 수행
# -> 얼마나 좋은 가중치 (weights) 인지 확인
minMisclassifications = 10000 # 적게 틀린 갯수 비교 위해 큰 값으로 잡음

# 반복 (iteration) 횟수 초기화
numIteration = 1000
```

```python
# 5. pocket 알고리즘 학습

# 최적의 가중치를 저장할 변수 설정
best_weights = np.zeros((3, 1)) # weight 담아둘 pocket 필요 placeholder 같은 것

for iteration in range(numIteration):
    iteration += 1
    
    misClassifications = 0 # 몇 개 틀렸는지 카운트

    for i in range(0, len(X_train)): # 0부터 999까지
        # 전체 데이터에서 하나씩 읽어옴
        currentX = X_train[i].reshape(-1, X_train.shape[1]) # 차원을 맞춰주는 것 
        currentY = y[i]

         # 각 데이터에 대해 w^T x 연산 수행
        wTx = np.dot(currentX, weights)[0][0]
        
        # PLA 알고리즘과 동일
        # -> 오분류의 기준이 퍼셉트론 (perceptron) 과 동일
        # -> 차이가 있다면 가중치 (learning rate) 가 입력에 곱해지는 것
        if currentY == 1 and wTx < 0: # currentY는 f(x), f(x)는 1이고 h(x)가 0보다 작은 것  (h(x)==-1) != (f(X)==1)
            misClassifications += 1 # 틀린 데이터 나올 때마다 하나씩 올려줌
            weights = weights + learningRate * np.transpose(currentX) # 실제 y가 1이기 때문에 +

        elif currentY == 0 and wTx > 0: # 원래 y, f는 -1 여기서는 0으로 
            misClassifications += 1
            weights = weights - learningRate * np.transpose(currentX) 
    
    # 최적의 가중치 확인 및 저장
    if misClassifications < minMisclassifications:
        minMisclassifications = misClassifications
        
        print('Keep best weights: {}'.format(minMisclassifications))
        best_weights = weights
    
    if iteration % 100 == 0:
        print("Iteration {}, Misclassifications {}".format(iteration, misClassifications))
```

```python
# 6. 학습 결과 확인

# 우리는 이전 셀 학습에서 minMisclassifications 에 가장 적게 틀린
print("Minimum Misclassifications : {}".format(minMisclassifications) )

print('Final weight is: {}'.format(weights.transpose()) )

print('Best weight is: {}'.format(best_weights.transpose()) )

# accuracy: (전체 데이터 수 - 틀린 데이터 수) / 전체 데이터 수
accuracy = ( X_train.shape[0] - misClassifications) / X_train.shape[0]

print("Best Accuracy of PocketAlgorithm is: {}".format(accuracy))
```

```python
# 7. 학습 결과 시각화

# 그려질 그래프의 두 축의 최대, 최소값 설정
max_x = np.max(X[:,0])
min_x = np.min(X[:,0])

max_y = np.max(X[:,1])
min_y = np.min(X[:,1])

def plot_hyperplane(X, Y, weights):
    
    # line 의 기울기 및 절편 계산
    slope = -weights[1] / weights[2]
    intercept = -weights[0] / weights[2]

    # line 생성
    x_hyperplane = np.linspace(-2, 3, 10)
    y_hyperplane = slope * x_hyperplane + intercept
    
    # 그래프 그리기
    plt.figure(figsize=(8,6))
    plt.scatter(X[:,0], X[:,1], c=Y, cmap=plt.cm.Spectral)

    plt.plot(x_hyperplane, y_hyperplane, '-')
    plt.title("Pocket algorithm on Half moon dataset")
    plt.xlabel("x_1")
    plt.ylabel("x_2")

    plt.xlim((min_x - 0.5, max_x + 0.5)) # 그래프가 너무 넓게 퍼지지 않도록 limit 걸어둠
    plt.ylim((min_y - 0.5, max_y + 0.5))
    
    plt.show()

# Final weights 로 line 그리기    
# plot_hyperplane(X, y, weights)

# Best weights 로 line 그리기
plot_hyperplane(X, y, best_weights)
```
