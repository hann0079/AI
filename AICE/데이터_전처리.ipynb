```python
결측치 N/A, Null, NaN
데이터 분석할 때 반드시 처리해야 함

결측치 처리 특징
1. 실무자의 견해가 많이 반영됨
2. 시간이 많이 투자되어야 함
3. 결측치를 제거하는 것은 가장 쉽게 처리하는 방법 - 데이터 손실
4. 데이터 편향 발생 (결측치를 단순 대체하는 경우)

결측치 처리 범위
1. Listwise - 결측치가 하나라도 존재하면 행을 삭제 - 정보 손실 최대
2. Pairwise - 모든 칼럼이 결측치면 행을 삭제, 결측치가 있는 칼럼의 값만 삭제 - 행마다 달라짐

결측치 처리 방법
1. 제거하기 -  데이터 손실이 발생
2. 채우기 - 값 대체하기 - 결측치를 나머지 값의 평균화 값으로 대체 (수치형 - (평균, 중앙값), 범주형 - 최빈값)
                        - 모든 결측치가 동일한 값을 가질 수 있음
          - 예측하기 - 모두 동일한 값으로 채우는 것이 아니라 주변값과의 상관관계, 예측모델을 통해 채우기
                     - 실제값은 아니기 때문에 채워진 값에 의해 평균, 중앙값, 상관관계가 영향을 받음
   - 데이터 손실 없이 결측치를 빠르게 채울 수 있음

python 결측치 처리 방법
1. fillna - 지정한 값으로 결측치를 채움 (슬라이싱하여 특정 열에 대해서만 채우는 것도 가능)
2. interpolate - 결측치를 같은 간격으로 채움 (선형적인 데이터 linear로 주변값 이용하여 같은 간격으로 채워짐)
3. dropna - 결측치가 있는 레코드를 제거 Listwise, 매개변수를 all로 주면 Pairwise 방식 사용

결측치 처리 주의사항
copy 메소드를 사용하지 않으면 주소값을 복사하기 때문에 원본 값이 변경됨.
원본을 보전하면서 데이터를 보정하려면 copy 메소드를 사용해야 함!
원본 데이터를 copy 해놓아야 roll back 가능 
중간 중간 원본 데이터 copy하기
```

```python
Q1와 Q3가 거리의 1.5배가 넘어가는 값을 Outlier 라고 합니다.
이 값들은 이상치로써 일반적으로 제거 또는 변경하여 데이터를 분석,학습 합니다.
단, 이상치 분석시에는 제거하지 않습니다.
```

```python
#이상치를 제거하는 함수 만들기
def removeOutliers(x, column): # 매개변수는 데이터프레임, 컬럼명
    # Q1, Q3구하기
    q1 = x[column].quantile(0.25) #quantile 사용하여 25%에 해당하는 값을 가져올 수 있음
    q3 = x[column].quantile(0.75) # 3사분위 값 구해짐
    
    # 1.5 * IQR(Q3 - Q1)
    iqr = 1.5 * (q3 - q1) # 1사분위와 3사분위 사이의 간격 구함 거기에 1.5를 곱하여 IQR 값 구함
    
    # 이상치를 제거
    y=x[(x[column] < (q3 + iqr)) & (x[column] > (q1 - iqr))] 
    # 1.5 배 넘어가지 않는 정상적인 값만 남김 y 데이터프레임에 넣어서 리턴해줌
    
    return(y)
```

```python
# 이상치를 변경하는 함수 만들기 - 대치하기 방법 사용
def changeOutliers(data, column):
    x=data.copy()
    # Q1, Q3구하기
    q1 = x[column].quantile(0.25)
    q3 = x[column].quantile(0.75)
    
    # 1.5 * IQR(Q3 - Q1)
    iqr = 1.5 * (q3 - q1)
    
    #이상치 대체값 설정하기
    Min = 0 # 정상 최소치보다 작은 데이터는 min값으로 대체
    if (q1 - iqr) > 0 : Min=(q1 - iqr)
        
    Max = q3 + iqr # 정상 최대치보다 큰 데이터는 max값으로 대체
    
    # 이상치를 변경
    # X의 값을 직졉 변경해도 되지만 Pyhon Warning이 나오기 떄문에 인덱스를 이용
    x.loc[(x[column] > Max), column]= Max # 최대값보다 크면 max로 대치
    x.loc[(x[column] < Min), column]= Min # 최소보다 작으면 min으로 대치
    
    return(x)
```

```python
#연속형 데이터의 이상치 변경하기
cust_data=changeOutliers(cust, 'avg_bill')
cust_data.describe()
```
