```python
1. Matplotlib 라이브러리
1-1. Matplotlib 개요
Matplotlib는 파이썬에서 데이터를 차트나 플롯(Plot)으로 그려주는 가장 많이 사용되는 데이터 시각화(Data Visualization) 패키지
라인 plot, 바 차트, 파이차트, 히스토그램, Box Plot, Scatter Plot 등을 비롯하여 다양한 차트와 plot 스타일을 지원합니다.
웹페이지(https://matplotlib.org/index.html) 에서 다양한 샘플 차트를 볼 수 있습니다.
```

```python
선 그래프 (Line Plot) 그리기
plt.plot(data) - 파라미터는 x,y 데이터
수치의 변화를 선으로 이어서 그려줌 - 시간에 따른 데이터의 변화 추세를 볼 때 사용

산점도 그리기 
plt.scatter(x,y) - plt.plot() 단계에서 대신 사용하는 것
키와 몸무게 같은 두 값 간의 관계를 표현
두 값이 양의 상관관계인지, 음의 상관관계인지 파악
X와 Y값을 지정해야 함

히스토그램 그리기
plt.hist(x) - plt.plot() 단계에서 대신 사용
수치형 데이터의 분포를 나타냄
빈도, 빈도밀도, 확률 등의 분포를 그릴 때 사용
구간 개수(bins)을 정해주지 않으면 10개 구간으로 나누어 그래프를 그림

박스 그래프 그리기
plt.boxplot(x) - plt.plot() 대신
수치적 자료를 표현하는 그래프
수치적 자료로부터 얻어낸 통계량인 5가지 요약 수치를 가지고 그래프를 만듦
5가지 요약 수치는 최소, 1사분위, 2사분위, 3사분위, 최대
박스의 아래 선은 최소, 박스 아래부분이 1사분위, 노란색 선은 중앙값, 선 위 끝은 최대 , 선 이외 점은 이상치를 넘어가는 데이터

막대 그래프 그리기
plt.bar(x, height) - plt.plot() 대신
범주형 데이터의 수치를 요약해서 보여줌
일반적으로 가로, 세로, 누적, 그룹화된 막대 그래프를 사용
```

```python
%matplotlib inline 
# %를 붙여서 주피터 노트북 명령어 -> matplotlib 명령어 차트를 바로 표시해달라는 말
import matplotlib.pyplot as plt
```

```python
1-2.Matplotlib 사용법
Matplotlib를 사용하여 시각화를 하기 위해서는 아래 3가지 코드를 기억해야 합니다.
아래를 단계별로 수행
1. plt.figure() - 시각화 그림을 표시할 영역 지정
2. plt.plot() - 시각화 차트 및 값 지정
3. plt.show() - 시각화 명령어
```

```python
#Matplotlib를 사용하여 간단한 차트를 그리기
plt.figure()
plt.plot([1,2,3], [100,120,110]) # x,y로 점이 생김
plt.show()
```

```python
차트 크기 키우기
차트의 크기는 그림을 표시할 영역의 크기를 키움으로 해결 할 수 있습니다.
figsize 옵션을 사용해서 조정 합니다. ( 기본 값은 (6,4) )
```

```python
plt.figure(figsize=(16,5))
plt.plot([1,2,3], [100,120,110])
plt.show()
```

```python
# Pandas에서 선그래프 그리기
#cust_data.csv 파일을 DataFrame으로 불러오기
import pandas as pd
df=pd.read_csv("cust_data.csv")
df.head()
```

```python
#불러온 데이터를 사용하여 선그래프를 그리기
df.plot() # df에서 숫자형 데이터를 인덱스 순서로 라인차트를 모두 그려줌 (크게 의미가 없다)
plt.show()
```

```python
# 하나의 열에 대해서만 시각화 하고 싶으면 컬럼 넣어주면 됨
#avg_bill만으로 그래프 그리기
df['avg_bill'].plot(figsize=(50,30))
plt.show()
```

```python
2-2. 산점도 그리기
산점도는 키와 몸무게 같은 두 값 간의 관계를 표현 합니다. - 데이터 분포 확인
두 값이 양의 상관관계인지 음의 상관관계인지를 파악할 수 있습니다.
산점도 위에 회귀선을 하나 그리면 하나의 값에 따라 다른 값이 어떻게 변하는지를 예측 할 수도 있습니다.
plt.scatter()를 사용하고 X와 Y값을 지정해야 합니다.
```

```python
#avg_bill, age간의 관계를 알아보기 위해 산점도 그리기
plt.figure(figsize=(16,6))
plt.scatter(y=df["avg_bill"], x=df["age"])
plt.show()
```

```python
2-3. 히스토그램 그리기
히스토그램은 수치형 데이터의 분포를 나타 냅니다.
흔히 빈도, 빈도밀도, 확율 등의 분포를 그릴때 사용합니다.
plt.hist()를 사용하며, 구간값(bins)을 정해주지 않으면 10개 구간으로 나누어 그래프를 그려 줍니다
```

```python
#A_bill에 대한 빈도를 10개 구간으로 그리기
plt.figure()
plt.hist(df["A_bill"])
plt.show()
```

```python
#age에 대한 빈도를 20개 구간으로 그리기
plt.figure()
plt.hist(df["age"],bins=20)
plt.show()
```

```python
2-4. 박스 그래프 그리기
수치적 자료를 표현하는 그래프입니다.
수치적 자료로부터 얻어낸 통계량인 5가지 요약 수치(최소값, 제 1사분위값, 제 2사분위값, 제 3사분위값, 최대값)를 가지고 그리프를 만듭니다.
plt.boxplot() 명령어를 사용합니다
```

```python
#임의의 값 리스트를 생성하고, 생성한 리스트를 사용하여 박스 그래프 그리기
x=[5, 3, 7, 10, 9, 5, 3.5, 8,6]
# 세로선의 하단 최소, 상단 최대, 75% 값이 8, 25% 값이 5, 중위값이 6임을 알 수 있음 
plt.boxplot(x=x)
plt.show()
```

```python
test = [5, 3, 7, 10, 9, 5, 3.5, 8,6]
print(sum(test)/9)
print(sorted(test)[4]) # 박스 그래프의 중위값은 정렬 시 중간 위치에 있는 값
```

```python
Pandas에서 가로 box plot 그리기
by는 Group화 할 값(컬럼), column은 박스 그래프로 나타낼 값(컬럼)을 넣어주면 됩니다.
```

```python
#나이대별 총이용금액 분포를 박스 그래프로 그리기
df.boxplot(by="by_age", column="avg_bill", figsize=(16,8)) 
plt.show()
```

```python
2-5. 막대 그래프(barplot)
막대 그래프는 범주형 데이터의 수치를 요약해서 보여주니다.
일반적으로 가로,세로,누적,그룹화 된 막대 그래프 등을 사용하며, plt.plot()대신 plt.bar() 명령어를 사용 할수있습니다.
```

```python
#임의의 값 리스트를 생성하고, 생성한 리스트를 사용하여 막대 그래프 그리기
y=[5, 3, 7, 10, 9, 5, 3.5, 8]
x=list(range(len(y))) # y의 길이만큼 인덱스를 만들어줌 x를

plt.figure()
plt.bar(x, y)
plt.show()
```

```python
pandas내 pivot_table 함수를 사용하면 쉽게 수치들에 대한 mean/sum/max등을 구할 수 있습니다.
pivot_table 을 배운 이유가 막대그래프 그리기 위해서임
```

```python
3-4. 마커활용하기 - line plot에서만 활용됨
```

```python
x=list(range(0,10))
y1=list(range(0,10))
y2=list(range(0,20,2))
y3=list(range(0,40,4))
```

```python
#(x,y1)(x,y2)(x,y3)의 선그래프를 그리기
plt.plot(x,y1,x,y2, x, y3) # x를 동일하게, y를 다르게 넣어주는 방법
plt.show()
```

```python
마커를 변경 하려면, 각 x,y값 뒤에 원하는 스타일 문자열을 넣어줍니다.
스타일 문자열은 색깔(color), 마커(marker), 선 종류(line style)의 순서로 지정하고,
만약 이 중 일부가 생략되면 기본 값이 적용된다.
색, 마커, 선 종류에 대한 설명은 plot의 도움말을 참고하도록 합니다.
```

```python
#3개의 선을 각기 다른 색과 마커, 선을 사용해 표현하기
plt.plot(x,y1,'r--', x,y2,'bs'  ,x, y3, 'g^:') # 마커 지정하려면 모든 x,y에 지정 (생략은 마커안에서 생략임) bs가 선 종류 생략인 것
plt.show()
```

```python
3-5. 여러개의 차트 한번에 그리기
지금까지는 하나의 Plot에 여러개의 라인을 추가 하였으나, 한번에 여러개의 plot을 그리는것도 가능합니다.
plt.show를 사용하여 그래프를 출력하기 전에 여러개의 figure를 열면 여러개의 그래프를 한번에 만들 수 있습니다.
```

```python
#(x,y), (x,y2)를 두개의 그래프로 동시에 그리기
plt.figure()
plt.plot(x,y1)
plt.figure()
plt.plot(x,y2)

plt.show()
```

```python
subplot을 사용하여 원하는 배치를 해서 한번에 출력하는것도 가능 합니다.
이때 subplot에 넘겨주는 값은 3개의 숫자로 앞에서부터 차례대로 행/열/순서 배치순서가 됩니다.

plt.subplot(221)는 2행 2열 배치에 첫번째 영역이 됩니다. 
plt.subplot(222)는 2행 2열 배치에 두번째 영역이 됩니다.
plt.subplot(212)는 2행 1열 배치의 두번째 영역 즉, 하단 긴 영역이 됩니다.
```

```python
5. Seaborn 목적별 차트 그리기
5-1. scatterplot
앞서 배웠듯 산점도는 두 값간의 상관관계를 확인해 볼 수 있습니다.
```

```python
#avg_bill과 age간의 상관관계를 확인해 봅시다.
sns.scatterplot(x='age', y='avg_bill', data=df) # x,y로 데이터의 칼럼명을 준다
plt.show()
# x,y 이름이 자동으로 들어가고 테두리 있음
```

```python
# matplotlib과의 비교
plt.scatter(x='age', y='avg_bill', data=df) # x,y로 데이터의 칼럼명을 준다
plt.show()
```

```python
5-2. catplot
catplot은 색상(hue)과 열(col) 또는 행(row) 등을 동시에 사용하여 3 개 이상의 카테고리 값에 의한 분포 변화를 보여줍니다.
```

```python
#age와 avg_bill간의 관계를 class별로 확인하기
sns.catplot(x='age', y='avg_bill',data=df ,col="class", col_wrap=2) # col wrap 으로 2개의 열로 패키징함
plt.show()
# C 클래스는 많은 사람이 요금을 적게 내고 있음을 알 수 있음 -> 낮은 고객 등급이겠군
```

```python
5-3. lmplot
lmplot은 산점도에 회귀선을 그려주어 하나의 값에 따라 다른 값이 어떻게 변하는지를 예측 할수 있습니다.
```

```python
#lmplot을 사용하여 avg_bill과 B_bill의 상관관계를 확인하기
plt.figure(figsize=(10,5))
sns.lmplot(x='avg_bill', y='B_bill', data=df,line_kws={'color': 'red'}) # line 지정 안하면 같은 색깔로 나옴
plt.show()
```

```python
#hue값으로 성별에 따라 어떻게 달라지는지 구분하기 
sns.lmplot(x='avg_bill', y='B_bill', data=df, hue='sex') # hue로 성별에 따라 두 개의 색깔로 나타나게 함
plt.show()
```

```python
5-4. countplot
항목 별 갯수를 확인 할 때 사용합니다.
히스토그램과 countplot은 두 그래프 모두 항목의 갯수분포를 확인하는 비슷한 그래프로 보이지만
히스토그램은 연속형 변수만 가능하고, coutplot은 범주형 변수에도 적용이 가능하다는 차이가 있습니다.
```

```python
#나이대별 bill_rating분포를 확인하기
plt.figure(figsize=(10,5))
sns.countplot(x="by_age", hue="bill_rating", data=df)
plt.legend(loc='upper right') # legend 수정 matplot에서 사용되는 것 똑같이 사용 가능
plt.show()
```

```python
#연속형 변수인 age로 count plot과 비슷한 그래프 그리기 
plt.figure(figsize=(10,5))
sns.histplot(x="age",bins=20, hue="bill_rating",data=df, multiple='dodge', shrink=0.8) 
# multiple을 dodge로 변경하여 쪼갬 shrink로 간격을 줌
plt.show()
```

```python
# Y축 값을 지정하면 가로 막대 그래프로 표현됩니다.
#성별 고객등급 분포를 가로 막대 그래프로 확인하기
sns.countplot(y="class", hue="sex", data=df)
plt.show()
```

```python
그래프 색깔 바꾸기
palette값을 설정하면 원하는 톤의 그래프를 그릴 수 있습니다.
더 많은 palette색상은 담은을 참고하세요. 
https://seaborn.pydata.org/tutorial/color_palettes.html
```

```python
#palette값을 spring로 설정하여 색을 변경하기
sns.countplot(y="class", hue="sex", data=df, palette='spring')
plt.show()
```

```python
5-5. jointplot
산점도와 countplot을 한번에 보여주어 데이터의 분포와 상관관계를 한번에 볼 수 있습니다.
```

```python
#jointplot을 사용하여 avg_bill과 age간의 관계 확인하기
sns.jointplot(x="avg_bill", y="age", data=df)
plt.show()
```

```python
# kind 값을 사용하면 산점도 영역의 그래프를 다른 그래프로 변경 할 수 있습니다
#kind 값을 변경하여 다양한 그래프를 그리기
sns.jointplot(x="avg_bill", y="age", data=df, kind='hex') # hex로 육각형 모양의 점으로 -> 어두우면 몰려있음
plt.show()
```

```python
5-6. Heatmap (상관관계)
상관관계를 확인하기 위해서 주의해야 할 점은 모든 데이터가 'continous'해야 한다는 점입니다. # 수치형 데이터여야 한다는 것
corr()은 상관관계를 측정 할 수 있는 함수입니다.
```

```python
df.corr()
# 각 컬럼별로 어떤 상관관계 있는지 수치로 나타남
# 1에 가까울수록 강한 양의 상관관계, -1에 가까울수록 강한 음의 상관관계, 0에 가까울수록 상관관계가 없다
```

```python
heatmap은 열을 의미하는 heat와 지도를 뜻하는 map을 합친 단어입니다.
데이터의 배열을 생삭으로 표현해 주는 그래프입니다.
heatmap을 사용하면 두개의 카테고리 값에 대한 값변화를 한 눈에 알기 쉽기 때문 두 값의 상관관계를 나타낼때 주로 사용 합니다.
```

```python
# '-' 부호 출력 경고를 제거 하기 위한 코드
plt.rc('axes', unicode_minus=False)
```

```python
#컬럼별 상과관계를 heatmap 그래프로 그리기
sns.heatmap(df.corr())
plt.show()
# 밝을수록 양, 어두울수록 음의 상관관계, 자주색일수록 상관관계 없다
# 나이, 평균 요금이 음의 상관관계 -> 양의 상관관계, 음의 상관관계에서 음의 상관관계 영향이 강했던 거임
```
